The apical junctions form a geometric graph that can be generated by `graph-tool`.Let's first generate the vertices as points in 3D with cylindrical coordinates$(\rho, \theta, z)$. We assume the apical part of the leg epithelium is initialya cylinder of radius $\rho_0$ and length $L=4\rho_0$ caped by two hemi-spheres(like _E. coli_ or _S. pombe_ ;-). The surface of the epithelium is given by $2\pi\rho_0 L = 8\pi \rho_0^2$. 

For a given average distance between vertices, $\lambda_0$, an hexagonal cell will occupy a surface $s_c = \frac{3\sqrt 3}{2} {\lambda_0}^2 / 2$. There will be around $N = S/s_c$ vertices in the model.
