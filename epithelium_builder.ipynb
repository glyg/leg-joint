The apical junctions form a geometric graph that can be generated by `graph-tool`.Let's first generate the vertices as points in 3D with cylindrical coordinates$(\rho, \theta, z)$. We assume the apical part of the leg epithelium is initialya cylinder of radius $\rho_0$ and length $L$ caped by two hemi-spheres(like _E. coli_ or _S. pombe_ ;-). The surface of the epithelium is given by $2\pi\rho_0 L + 4\pi\rho_0^2$. With $ L = 2 * \rho_0 $, this boils down to:$S = 8\pi \rho_0^2$. 

```python
rho_0, rho_noise = 60, 0.01surface = 8 * np.pi * rho_0 **2print r'S = %.0f µm²' % surface
```

For a given average distance between vertices, $\overline\lambda$, there will be around $N = \frac{S}{{\overline\lambda}^2}$ vertices in the model, half of which will belong to the cylindrical part, with one of the remaining quarters for each cap.

```python
lambda_avg = 6 #µmnum_vertices = 2 * (np.int(surface / (lambda_avg**2)) // 2)print 'there will be  %i vertices' % num_vertices
```
