```python
def new_2pannels(eptm):
    
    eptm.update_rhotheta()
    
    colors = eptm.rhos.copy()
    eptm.graph.set_vertex_filter(eptm.is_cell_vert)
    colors.a -= 5#colors.fa.min()
    colors.a /= 25#colors.fa.max()
    
    eptm.graph.set_vertex_filter(None)
    
    rot = 45
    rot = -(rot + 90) * np.pi / 180
    eptm.rotate(rot)
    lj.local_slice(eptm, theta_amp=np.pi/5, theta_c=0, zed_amp=10, zed_c=0)
    
    fig, (ax_zs, ax_3d) = plt.subplots(1, 2, figsize=(12, 4), sharey=True)
    
    ax_zs = lj.plot_eptm_generic(eptm, eptm.zeds, eptm.proj_sigma(), ax=ax_zs,
                                 cell_kwargs={'c_text':False,
                                              'cell_colors':colors,
                                              #'c':'k',
                                              'alpha':0.4, 'normalize':False},
                                 edge_kwargs={'j_text':False,
                                              'c':'k',
                                              'lw':0.4,
                                              'alpha':0.5})
    
    
    eptm.rotate(- rot)
    
    rhos, thetas, zeds = eptm.rhos, eptm.thetas, eptm.zeds
    z_angle=np.pi/24
    d_theta = - 6 * np.pi / 7
    
    pseudo_x = eptm.graph.new_vertex_property('float')
    pseudo_y = eptm.graph.new_vertex_property('float')
    pseudo_x.a = zeds.a * np.cos(z_angle) - rhos.a * np.cos(
        thetas.a + d_theta) * np.sin(z_angle)
    pseudo_y.a = rhos.a * np.sin(thetas.a + d_theta)
    
    edge_alpha = eptm.dzeds.copy()
    
    v_depth = eptm.zeds.copy()
    v_depth.a = rhos.a * (1 - np.cos(thetas.a + d_theta))
    
    v_prop = v_depth
    
    alpha0 = 0.
    edge_alpha.a = np.array([(v_prop[s] + v_prop[t])
                             for s, t in eptm.graph.edges()])
    
    edge_alpha.a -= edge_alpha.a.min()
    edge_alpha.a += alpha0 * edge_alpha.a.max()
    
    edge_alpha.a /= edge_alpha.a.max()
    
    
    
    ax_3d = lj.plot_edges_generic(eptm, pseudo_x, pseudo_y, ax=ax_3d, efilt=None,
                                  edge_alpha=edge_alpha,
                                  edge_color=edge_alpha,
                                  j_text=False, 
                                  c='g',
                                  lw=0.6)
    
    ax_3d.plot([-40, -40], [-15, -20], '-k', lw=2)
    ax_zs.plot([-20, -20], [-15, -20], '-k', lw=2)
    
    for ax in (ax_zs, ax_3d):
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_frame_on(False)
        ax.set_ylim(-35, 35)
        
    plt.savefig(os.path.join(eptm.paths['svg'], 'two_pannels.svg'))
```

```python
sigmas = eptm.proj_sigma()

def cell_polygon(cell, step):
    jv_zeds = np.array([eptm.zeds[jv] - eptm.zeds[cell]
                        for jv in cell.out_neighbours()])
    jv_sigmas = np.array([sigmas[jv] - sigmas[cell]
                         for jv in cell.out_neighbours()])
    
    jv_rhos = np.hypot(jv_zeds, jv_sigmas)
    jv_thetas = np.arctan2(jv_zeds, jv_sigmas)
    jv_zeds = jv_zeds[np.argsort(jv_thetas)]
    jv_sigmas = jv_sigmas[np.argsort(jv_thetas)]
    sg_zs = np.vstack([jv_zeds, np.roll(jv_zeds, shift=1)]).T
    sg_ss = np.vstack([jv_sigmas, np.roll(jv_sigmas, shift=1)]).T
    lengths = np.hypot(sg_zs[:,0] - sg_zs[:,1],
                       sg_ss[:,0] - sg_ss[:,1])
    num_points = np.round(lengths / step)
    line_z = np.concatenate([np.linspace(*sg_z, num=n_pts)[:-1]
                             for sg_z, n_pts in zip(sg_zs, num_points)])
    line_s = np.concatenate([np.linspace(*sg_s, num=n_pts)[:-1]
                             for sg_s, n_pts in zip(sg_ss, num_points)])
    return line_z, line_s
```

```python
def ellipse(thetas, params):
    
    a, b, phi = params
    x = a * np.cos(thetas) # + xc
    y = b * np.sin(thetas + phi) # + yc
    return x, y

def residuals(params, data):
    
    x, y = data
    thetas = np.arctan2(y, x)

    fit_x, fit_y = ellipse(thetas, params)
    res_x = fit_x - x
    res_y = fit_y - y
    return np.hstack((res_y, res_x))
```

```python
def fit_ellipse(line_z, line_s):
   
    a = line_z.ptp()/2
    b = line_s.ptp()/2
    phi = 0

    thetas = np.linspace(0, 2*np.pi)
    params, r = sp.optimize.leastsq(residuals, 
                                    [a, b, phi,],
                                    [line_z, line_s])
    return params, r
```
