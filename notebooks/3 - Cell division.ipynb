## Abscission of a cell

```python
eptm.set_local_mask(None)
mother_cell = eptm.graph.vertex(45)
eptm.cells.prefered_area[mother_cell] /= 2.
eptm.cells.contractilities[mother_cell] /= 2.

for edge in eptm.cell_junctions(mother_cell):
    eptm.junctions.line_tensions[edge] /= (2*(1.5))

    
eptm.set_local_mask(mother_cell)
fig, axes = plt.subplots(1,2, figsize=(12,4))
plot_cells_sz(eptm, axes[0], text=False, 
              vfilt=eptm.is_local_vert,
              efilt=eptm.is_local_edge)

j= cell_division(eptm, mother_cell, verbose=True)

#jv0, jv1, mother_cell, daughter_cell = j
plot_cells_sz(eptm, axes[1], text=False,
              vfilt=eptm.is_local_vert,
              efilt=eptm.is_local_edge)
#print j
```
